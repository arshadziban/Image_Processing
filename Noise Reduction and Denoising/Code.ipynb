```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("noisy_image.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.axis('off') 
```

```python
denoised_image = cv2.GaussianBlur(image, (5, 5), 0)

# Display the denoised image
plt.subplot(1, 2, 2)
plt.title("Denoised Image")
plt.imshow(denoised_image, cmap='gray')
plt.show()
```

```python
color_denoised_image = cv2.applyColorMap(denoised_image, cv2.COLORMAP_JET)

# Display the denoised image in color
plt.subplot(1, 3, 2)
plt.title("Colored Denoised Image")
plt.imshow(color_denoised_image)
```

```python
# Apply Non-Local Means Denoising for noise reduction
denoised_image = cv2.fastNlMeansDenoising(image, None, h=30, templateWindowSize=7, searchWindowSize=21)

# Resize the denoised image for better quality (optional)
denoised_image = cv2.resize(denoised_image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
plt.subplot(1, 3, 3)
plt.title("Denoised Image (Grayscale)")
plt.imshow(denoised_image, cmap='gray')
plt.axis('off')
```
