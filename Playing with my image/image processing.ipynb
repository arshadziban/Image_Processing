```python
import cv2
import matplotlib.pyplot as plt

# read image
image_path = 'C:\\Users\\USER\\Desktop\\Programming\\Image processing\\ziban.jpg'  
image = cv2.imread(image_path)
```

```python
#resize image
resized_image = cv2.resize(image, (200, 200))
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
```

```python
#real color to gray color
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
#blur image
blurred_image = cv2.GaussianBlur(resized_image, (5, 5), 0)

# Display the blurred image
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image')
```

```python
# Using Canny edge detection
edges = cv2.Canny(resized_image, 100, 200)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
```

```python
# Flip the image horizontally (1), vertically (0), or both (-1)
flipped_image = cv2.flip(resized_image, 1)
plt.imshow(cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB))
plt.title('Flipped Image')
plt.axis('off')
plt.show()
```

```python
# Increase brightness by adding a constant to all pixel values
bright_image = cv2.convertScaleAbs(resized_image, alpha=1, beta=50)  # Alpha is for contrast, beta is for brightness
plt.imshow(cv2.cvtColor(bright_image, cv2.COLOR_BGR2RGB))
plt.title('Brightened Image')
```

```python

# Apply fastNlMeansDenoising for color images
denoised_image = cv2.fastNlMeansDenoisingColored(resized_image, None, 10, 10, 7, 21)
plt.imshow(cv2.cvtColor(denoised_image, cv2.COLOR_BGR2RGB))
plt.title('Denoised Image')
plt.axis('off')
plt.show()
```
